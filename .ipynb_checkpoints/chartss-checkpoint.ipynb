```python
import pandas as pd  
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("C:/Users/LENOVO/Downloads/Employee Sample Data.xlsx")
```

```python
Ethnicity = df['Ethnicity'].value_counts()  
plt.figure(figsize=(6, 6))
plt.pie(Ethnicity, 
        labels=Ethnicity.index, 
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops={'edgecolor': 'black'})  

plt.title('Ethnicity Distribution')

plt.show()


```

```python
Gender = df['Gender'].value_counts()  
plt.figure(figsize=(6, 6))
plt.pie(Gender, 
        labels=Gender.index, 
        autopct='%1.1f%%',
        startangle=90,
        colors=['#ff9999','#66b3ff'],
        wedgeprops={'edgecolor': 'black'})  

plt.title('Gender Distribution')

plt.show()
```

```python
Country = df['Country'].value_counts()  
plt.figure(figsize=(6, 6))
plt.pie(Country, 
        labels=Country.index, 
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops={'edgecolor': 'black'})  

plt.title('Country Distribution')

plt.show()
```

```python

ethnicity_salary = df.groupby('Ethnicity')['Annual Salary'].mean() 

ethnicity_salary.plot(kind='bar')

plt.title('Average Annual Salary by Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Average Annual Salary')

plt.show()
```

```python
import numpy as np

minn = df['Age'].min()
maxx = df['Age'].max()

bins = np.arange(minn, maxx + 5, 5)  
labels = [f'{i}-{i+4}' for i in bins[:-1]] 

df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_salary = df.groupby('age_group')['Annual Salary'].mean()

age_salary.plot(kind='bar')

plt.title('Average Salary by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Annual Salary')

plt.show()
```

```python
Gender_salary = df.groupby('Gender')['Annual Salary'].mean() 

Gender_salary.plot(kind='bar')

plt.title('Average Annual Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Annual Salary')

plt.show()
```

```python
Age_Bonus = df.groupby('Age')['Bonus %'].mean() 

Age_Bonus.plot(kind='bar', figsize=(10, 6))

plt.title('Average Bonus by Age')
plt.xlabel('Age')
plt.ylabel('Bonus')

plt.show()
```

```python
Gender_Bonus = df.groupby('Gender')['Bonus %'].mean() 

Gender_Bonus.plot(kind='bar')

plt.title('Average Bonus by Gender')
plt.xlabel('Gender')
plt.ylabel('Bonus')

plt.show()
```

```python
ethnicity_Bonus = df.groupby('Ethnicity')['Bonus %'].mean() 

ethnicity_Bonus.plot(kind='bar')

plt.title('Average Bonus by Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Bonus')

plt.show()
```
